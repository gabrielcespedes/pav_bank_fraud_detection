```python
variables = ['income', 'customer_age', 'payment_type', 'bank_branch_count_8w', 'employment_status', 'credit_risk_score', 'email_is_free', 'housing_status', 'session_length_in_minutes', 'device_os']
```

```python
ncols = 2
nrows = (len(variables) + 1) // ncols

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(12, 10))
axes = axes.flatten()

for i, var in enumerate(variables):
    sns.histplot(data=df, x=var, hue='fraud_bool', ax=axes[i])
    axes[i].set_title(f'DistribuciÃ³n de {var}')
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Frecuencia')

# Oculta ejes no usados si la cuadrÃ­cula es impar
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
df_negativa = df[df['fraud_bool'] == 0]
df_aleatorio = df_negativa.sample(n = 33372, random_state = 42)
df_positiva = df[df['fraud_bool'] == 1]
df_aleatorio = pd.concat([df_aleatorio, df_positiva], axis = 0)
df_aleatorio["fraud_bool"].value_counts()
```
